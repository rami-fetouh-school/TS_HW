```python
#Import libraries
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os
```

```python
#Read the files as dataframes
calendar = pd.read_csv('/content/drive/MyDrive/calendar.csv')
sales_train_validation = pd.read_csv('/content/drive/MyDrive/sales_train_evaluation.csv')
sell_prices = pd.read_csv('/content/drive/MyDrive/sell_prices.csv')
```

1. Which state has the highest sales? (5 points)

```python
#File: sales_train_validation. Group by state_id and sum the sales row
a=sales_train_validation.groupby('state_id').sum()
a
a.columns
a['sum'] = a.sum(axis=1)
a.sort_values(by = 'sum',ascending=False)

#Answer: CA
```

2. Which department has the highest sales? (5 points)

```python
#File: sales_train_validation. Group by dept_id and sum the sales row
a=sales_train_validation.groupby('dept_id').sum()
a
a.columns
a['sum'] = a.sum(axis=1)
a.sort_values(by = 'sum',ascending=False)

#Answer: FOODS_3
```

3. Which department has the highest number of products? (5 points)

```python
#Group by dept_id and then count unique item_id
a = sales_train_validation.groupby('dept_id')['item_id'].nunique()
a.sort_values(ascending=False)

#Answer: FOODS_3
```

4. Which department has the highest mean price? (5 points)

```python
#File: sell_prices. Split item_id. Group by dept_id and find average
sell_prices['dept_id'] = sell_prices.item_id.str.split('_', expand=True)[0] + '_' + sell_prices.item_id.str.split('_', expand=True)[1]

a=sell_prices.groupby(['dept_id']).mean()
a

#Answer: Household_2
```

5. Which is the best performing store? (5 points)

```python
#File: sales_train_validation. Group by state_id and sum the sales row
a=sales_train_validation.groupby('store_id').sum()
a.columns
a['sum'] = a.sum(axis=1)
a.sort_values(by = 'sum',ascending=False)

#Answer: CA_3
```

6. Which month had the highest sales? (5 points)

```python
#Match total sales train validation to calendar months by 'd'
a=pd.DataFrame(sales_train_validation.sum()[6:])
a['d'] = a.index
b = calendar[['d','month']]
a = a.merge(b,on='d', how='left')

#Find sum
a.groupby('month').sum()[0].sort_values(ascending=False)

#Answer: March
```

7. Which weekday do people prefer to grocery shopping in different states? (3 answers = 15 points)

```python
#California: Answer: Sunday
sales_train_validation_CA = sales_train_validation.loc[sales_train_validation['state_id'] == 'CA']
a=pd.DataFrame(sales_train_validation_CA.sum()[6:])
a['d'] = a.index
b = calendar[['d','weekday']]
a = a.merge(b,on='d', how='left')

print("CA:")
print(a.groupby('weekday').sum()[0])

#Texas: Answer: Sunday
sales_train_validation_TX = sales_train_validation.loc[sales_train_validation['state_id'] == 'TX']
a=pd.DataFrame(sales_train_validation_TX.sum()[6:])
a['d'] = a.index
b = calendar[['d','weekday']]
a = a.merge(b,on='d', how='left')

print("TX:")
print(a.groupby('weekday').sum()[0])

#Wisconsin: Answer: Saturday
sales_train_validation_WI = sales_train_validation.loc[sales_train_validation['state_id'] == 'WI']
a=pd.DataFrame(sales_train_validation_WI.sum()[6:])
a['d'] = a.index
b = calendar[['d','weekday']]
a = a.merge(b,on='d', how='left')

print("WI:")
print(a.groupby('weekday').sum()[0])
```
